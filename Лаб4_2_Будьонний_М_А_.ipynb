```python
#Імпортуємо необхідні бібліотеки
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Реалізація функції koch_snowflake, яка генерує вершини для малювання сніжинки Коха заданого порядку та розміру.
def koch_snowflake(order, size):
    if order == 0:
        return np.array([[0, 0], [size / 2, 0], [size / 4, (np.sqrt(3) / 2) * size], [0, 0]])
    else:
        pts = koch_snowflake(order - 1, size)
        new_pts = []
        for i in range(len(pts) - 1):
            pt1 = pts[i]
            pt2 = pts[i + 1]
            dx = pt2[0] - pt1[0]
            dy = pt2[1] - pt1[1]
            new_pts.append(pt1)
            new_pts.append([pt1[0] + dx / 3, pt1[1] + dy / 3])
            new_pts.append([pt1[0] + dx / 2 + (dy * np.sqrt(3) / 6), pt1[1] + dy / 2 - (dx * np.sqrt(3) / 6)])
            new_pts.append([pt1[0] + 2 * dx / 3, pt1[1] + 2 * dy / 3])
        new_pts.append(pts[-1])
        return np.array(new_pts)
```

```python
# Введення рівня рекурсії та розміру
order = int(input("Введіть рівень рекурсії (рекомендовано від 1 до 5): "))
size = 300
```

```python
# Побудова крижинки Коха
snowflake = koch_snowflake(order, size)
plt.figure(figsize=(6, 6))
plt.plot(snowflake[:, 0], snowflake[:, 1], 'b')
plt.axis('equal')
plt.axis('off')
plt.show()
```

Отже, у висновку створення крижинки Коха зможе допомогти в розумінні принципів рекурсивних алгоритмів та геометричних об'єктів.  Наша задача демонструє, як прості правила можуть бути використані для створення складних та естетично приємних візуальних ефектів. Крім того, регулювання рівня рекурсії дозволяє відчути вплив кількості ітерацій на складність та красу крижинки, що підсилює розуміння алгоритму.
